# **ALARM CLOCK**

````python
import tkinter as tk
import datetime as dt
from tkinter import ttk
````

````python
# GUİ'nin oluşturulması

# 1.Adım:
root = tk.Tk()
root.title("Alarm Clock")

# Gün bilgisi alınır.
moment = dt.datetime.now()
day = dt.datetime.strftime(moment, '%d %B %Y %A')  
# %d:sayı değerli bir karakter dizisi olarak gün
# %B:ayın tam adı
# %Y:yılın dört haneli tam hali
# %A:hafta gününün tam adı

# Saat bilgisi alınır.
hour, minute, second = dt.datetime.strftime(moment, '%H %M %S').split()  

# Ana çerçeve oluşturulur.
main_frame = tk.Frame(root)
main_frame.pack(side="top", padx=20, pady=(30, 10), expand=True, fill="both")  # Yukarıda hizalanır ve boşluk eklenir

# Gün başlığı
label_day_text = tk.Label(main_frame, text="DATE", font=("Arial", 14, "bold"), fg="black")
label_day_text.pack(anchor='center', pady=(5, 2))

# Gün bilgisi kutucuğu
label_day_value = tk.Label(main_frame, text=day, font=("Arial", 12, "bold"), bg="steelblue", fg="white", width=25)
label_day_value.pack(anchor='center')



# 2. Adım:
# Saat için çerçeve oluşturulur.
frame_time = tk.Frame(main_frame)
frame_time.pack(anchor='center', pady=(20, 10))  # Ortaya hizalar

# Saat
label_hour_value = tk.Label(frame_time, text=hour, font=("Arial", 10, "bold"), bg="steelblue", fg="white", width=5)
label_hour_value.grid(row=0, column=0, padx=2)

colon1 = tk.Label(frame_time, text=":", font=("Arial", 10, "bold"), fg="black")
colon1.grid(row=0, column=1, padx=2)

# Dakika
label_minute_value = tk.Label(frame_time, text=minute, font=("Arial", 10, "bold"), bg="steelblue", fg="white", width=5)
label_minute_value.grid(row=0, column=2, padx=2)

colon2 = tk.Label(frame_time, text=":", font=("Arial", 10, "bold"), fg="black")
colon2.grid(row=0, column=3, padx=2)

# Saniye
label_second_value = tk.Label(frame_time, text=second, font=("Arial", 10, "bold"), bg="steelblue", fg="white", width=5)
label_second_value.grid(row=0, column=4, padx=2)



# 3. Adım:
frame_label = tk.Frame(main_frame)
frame_label.pack(anchor='center', pady=(10, 10))

# Başlık etiketi
label = tk.Label(frame_label, text="When to wake you up:", font=("Arial", 10, "bold"), fg="red")
label.grid(row=0, column=0, columnspan=2, pady=5)

# Kullanıcı gün seçimi
day_select = tk.Spinbox(frame_label, from_=1, to=31)
day_select.grid(row=1, column=0, padx=5, pady=5)

# Kullanıcı ay seçimi
def on_select(event):
    selected_item = month_select.get()

month_select = ttk.Combobox(frame_label, values=["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"])
month_select.set("Month?")
month_select.bind("<<ComboboxSelected>>", on_select)
month_select.grid(row=1, column=1, padx=5, pady=5)



# 4.Adım:
frame_hour = tk.Frame(main_frame)
frame_hour.pack(anchor='center', pady=(10,10))

# Kullanıcı saat seçimi
label_hour = tk.Label(frame_hour, text="Hour:", font=("Arial", 10, "bold"), fg="black")
label_hour.grid(row=0, column=0, columnspan=3, pady=5)

hour_select = tk.Spinbox(frame_hour, from_=0, to=23, width=2, font=("Arial", 10))
hour_select.grid(row=1, column=0, padx=2)

colon_label2 = tk.Label(frame_hour, text=":", font=("Arial", 10, "bold"))
colon_label2.grid(row=1, column=1, padx=2)

minute_select = tk.Spinbox(frame_hour, from_=0, to=59, width=2, font=("Arial", 10))
minute_select.grid(row=1, column=2, padx=2)

# Buton oluşturulur.
frame_button = tk.Frame(main_frame)
frame_button.pack(anchor='center', pady=(10, 10))

def show_message():
    label_message.config(text="Alarm Created!")

label_message = tk.Label(main_frame, text="", font=("Arial", 10, "italic"), fg="green")
label_message.pack(pady=5)

button = tk.Button(frame_button, text="Click Me", activebackground="green", activeforeground="white", command=show_message)
button.grid(row=2, column=0, columnspan=2, pady=10)



# Uyarı alarmının oluşturulması:

def warning_alarm():
    # Bulunula zaman alınır.
    current_time = dt.datetime.now()
    
    # Kullanıcıdan alınan değerlerdönüştürülerek değişkenlere atanır.
    selected_month = month_select.get()
    selected_day = int(day_select.get())
    selected_hour = int(hour_select.get())
    selected_minute = int(minute_select.get())
    
    # Ayın ismi sayısal değere dönüştürülür.
    month_number = dt.datetime.strptime(selected_month, "%B").month if selected_month != "Month?" else None
    
    # Alarm
    if (selected_month != "Month?" and 
        current_time.day == selected_day and
        current_time.hour == selected_hour and
        current_time.minute == selected_minute and
        current_time.month == month_number):
        print("Wake Up!!!")
        
    # Alarmı kontrol etmek için her saniye tekrar çağırır.
    root.after(10, warning_alarm)
    
warning_alarm()
        
# Pencere boyutu ayarlanır.
root.geometry("600x400")  
root.mainloop()
````

* `grid()` metodu, Tkinter'de widget'leri bir ızgara (grid) sistemi içinde konumlandırmamıza olanak tanır. Bu ızgara sistemi satır ve sütunlardan oluşur ve her bir hücre, row ve column parametreleriyle belirlenir. `grid()` metodunda kullanılan ana parametreler:


`row`: Widget'in hangi satırda yer alacağını belirler.


`column`: Widget'in hangi sütunda yer alacağını belirler.


`pady ve padx`: Widget'in etrafına dikey (pady) veya yatay (padx) boşluk ekler.


```python
# Satır 0 - Başlıklar
label_hour_text.grid(row=0, column=0)
label_minute_text.grid(row=0, column=1)
label_second_text.grid(row=0, column=2)

# Satır 1 - Değerler
label_hour_value.grid(row=1, column=0)
label_minute_value.grid(row=1, column=1)
label_second_value.grid(row=1, column=2)
```

* `pady` parametresi, bir widget (örneğin bir etiket, düğme veya çerçeve) için dikey boşluk eklemek amacıyla kullanılır. `pady`, widget`ın üst ve alt taraflarına eklenen boşluk miktarını belirler.

`pady` iki değer alabilir:


`Tek bir değer`: Örneğin, pady=10 yazıldığında, bu değer hem üst hem alt kısımdaki boşluğu 10 piksel olarak ayarlar.

`İki değer`: Örneğin, pady=(5, 10) yazıldığında, ilk değer üst boşluğu, ikinci değer ise alt boşluğu ayarlar. Bu durumda, üst boşluk 5 piksel, alt boşluk 10 piksel olacak şekilde ayarlanır.
